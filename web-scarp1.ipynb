```python
import requests
import bs4
import pandas as pd
```

```python
r=requests.get("https://www.flipkart.com/laptops/pr?sid=6bo,b5g&otracker=categorytree&fm=neo%2Fmerchandising&iid=M_d8a82938-bd09-473e-b328-528f16e29747_1_372UD5BXDFYS_MC.34WHNYFH5V2Y&otracker=hp_rich_navigation_8_1.navigationCard.RICH_NAVIGATION_Electronics~Laptop%2Band%2BDesktop_34WHNYFH5V2Y&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_8_L1_view-all&cid=34WHNYFH5V2Y")
```

```python
r.content
```

```python
scoup=bs4.BeautifulSoup(r.content)
```

```python
print(scoup.prettify())
```

```python
divs=scoup.find_all('div',attrs={'class':'_1AtVbE col-12-12'})
```

```python
len(divs)
```

```python
data=[]
for item in divs:
    laptopTitle=item.find('div',attrs={'class':'_4rR01T'})
    if laptopTitle is not None:
        laptopPrice=item.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
        laptopacprice=item.find('div',attrs={'class':'_3I9_wc _27UcVY'})
        rating=item.find('div',attrs={'class':'_3LWZlK'})
        if rating is not None:
            rating=rating.text
        else:
            rating=0    
        print(laptopTitle.text.__str__().split('-')[0].strip(), "\t",laptopPrice.text.__str__()[1:].replace(',',''))
        data.append([laptopTitle.text.__str__().split('-')[0].strip(),laptopPrice.text.__str__()[1:].replace(',',''),rating])
```

```python
df=pd.DataFrame(data,columns=['Laptop Name','Price','Rating'])
```
